```python
def load_json(file_path):
    import json
    with open(file_path, 'r') as f:
        return json.load(f)
    
def save_json(file_path, data):
    import json
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=4)

def save_txt(file_path, data):
    with open(file_path, 'w') as f:
        f.write(data)
```

```python
def scrap_save_case(case_id, one_liner, url, year):
    scraper = LegalCaseScraperFULL()
    try:
        case = scraper.scrape_case(url)
        case_text_list, case_metadata_list = case['text'], case['metadata']
        case_text_string = '\n'.join(case_text_list)
        case_text_metadata = '\n'.join(case_metadata_list)
        kc_one_liner = one_liner
        # write case as txt
        save_txt(f'./key_cases_scrapped_full/{year}-{case_id}.txt', case_text_string)
        save_txt(f'./key_cases_scrapped_full/{year}-{case_id}-metadata.txt', case_text_metadata)
        save_txt(f'./key_cases_scrapped_full/{year}-{case_id}-one-liner.txt', kc_one_liner)
    finally:
        scraper.close()
```

```python
def scrape_year(year):
    data = load_json(f'./key_cases_meta_data/kc_{year}_parsed.json')
    case_list = data['cases']
    for case in case_list:
        url = case['link']
        one_liner = case['case_text']
        case_id = url.split('=')[-1]
        print(f"Scraping {case_id}")
        scrap_save_case(case_id, one_liner, url, year)
        print(f"=====================")
```
