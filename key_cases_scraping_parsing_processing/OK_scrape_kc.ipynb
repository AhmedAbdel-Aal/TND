```python
from scrap_case import LegalCaseScraper
```

```python
def load_json(file_path):
    import json
    with open(file_path, 'r') as f:
        return json.load(f)
    
def save_json(file_path, data):
    import json
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=4)
```

```python
def scrap_save_case(case_id, meta_data, url, year):
    scraper = LegalCaseScraper()
    try:
        facts = scraper.scrape_facts(url)
        name = url.split('=')[-1]
        data = {
            'url': url,
            'name': name,
            'meta_data': meta_data,
            'facts': facts
        }
        # write facts as json
        save_json(f'./key_cases_scrapped/{year}-{case_id}.json', data)
    finally:
        scraper.close()
```

```python
def scrape_year(year):
    data = load_json(f'./key_cases_meta_data/kc_{year}_parsed.json')
    case_list = data['cases']
    for case in case_list:
        url = case['link']
        meta_data = case['case_text']
        case_id = url.split('=')[-1]
        print(f"Scraping {case_id}")
        scrap_save_case(case_id, meta_data, url, year)
        print(f"=====================")
```

```python
scrape_year('2024')
```
